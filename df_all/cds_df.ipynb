#### Dataframe Creation

```python
import pandas as pd
import numpy as np
import re
```

```python
# instantiate main dataframe, df

df = pd.DataFrame()

df['Dx ID'] = 0
df['Mx ID'] = 0
df['Dx Mx Association'] = 0
df['Dx Name'] = 0
#df['Dx Prevalence'] = 0 # we'll merge() this later
df['Mx Name'] = 0
df['Mx PPV'] = 0
df['Mx Sensitivity'] = 0
#df['Mx Import'] = 0 # we'll merge() this later
df['Specificity'] = 0
df['LR (+)'] = 0
df['LR (-)'] = 0

df
```

```python
# incorporate the associations between Dx and Mx
df_DxMx = pd.read_csv('dx_mx_associations.txt', sep='\t', names=['dx', 'mx'])
df_DxMx.head()
```

```python
# read Dx Mx file

f = open('dx_mx_associations.txt')
lines = f.readlines()
f.close

lines[:5]
```

```python
# group the Dx ID, Dx name, Dx prevalence, Mx ID, Mx name, Mx PPV, and Mx Sensitivity

m = re.compile(r'DX\s(\d+)\s+(.+)\s\sMX\s(\d)(\d)\s(\d+)\s(.+)')

dx_id = []
dx_name = []
mx_id = []
mx_name = []
mx_ppv = []
mx_sens = []

for i in range(len(lines)):
    dx_id.append((m.match(lines[i]).group(1)))
    dx_name.append((m.match(lines[i]).group(2)))
    mx_id.append((m.match(lines[i]).group(5)))
    mx_name.append((m.match(lines[i]).group(6)))
    mx_ppv.append((m.match(lines[i]).group(3)))
    mx_sens.append((m.match(lines[i]).group(4)))
```

```python
# combine Dx and Mx IDs for associations column

dx_mx = []

for i in range(len(dx_id)):
    dx_mx.append(dx_id[i] + ', ' + mx_id[i])
    
dx_mx[:5]
```

```python
# incorporate prevalence table
df_prev = pd.read_csv('dx_prev.txt', sep='\t')

# Grayson, I'm renaming these columns to make a merge simpler in a future step
df_prev.rename(index=str, columns={"id": "Dx ID", "prev": "Dx Prevalence"}, inplace=True)

# Grayson, I'm also coercing the index to a string to be consistent with the 
# df dataframe you created
df_prev['Dx ID'] = df_prev['Dx ID'].apply(str)
df_prev.head()
```

```python
# incorporate import numbers
f = open('mx.txt')
lines = f.readlines()
lines[:5]
```

```python
# group the Mx ID numbers and import numbers

m = re.compile(r'MX\s(\d+).+IM\s+(\d)')

mx_import_id = []
mx_import_num = []

for i in range(len(lines)):
    mx_import_id.append((m.match(lines[i]).group(1)))
    mx_import_num.append((m.match(lines[i]).group(2)))
```

```python
# create dataframe for imports
df_mx_import = pd.DataFrame()
df_mx_import['Mx ID'] = mx_import_id
df_mx_import['Import Num'] = mx_import_num
```

```python
df_mx_import.tail()
```

#### Add everything to df

```python
# put associations, Dx name, Mx name, PPV, and sensitivity into df

df['Dx ID'] = dx_id
df['Mx ID'] = mx_id
df['Dx Mx Association'] = dx_mx
df['Dx Name'] = dx_name
df['Mx Name'] = mx_name
df['Mx PPV'] = mx_ppv
df['Mx Sensitivity'] = mx_sens

df.head()
```

```python
# add import numbers
#for i in range(len(df)):
#    for j in range(len(df_mx_import)):
#        if df['Mx ID'][i] == mx_import_id[j]:
#            df['Mx Import'][i] = mx_import_num[j]

# Grayson, you can do a simple 'merge'
# (using set theory) rather than a double for-loop
# should be much faster

df = pd.merge(df, df_mx_import, on='Mx ID')
```

```python
# add prevalences
#for i in range(len(df)):
#    for j in range(len(df_prev)):
#        if int(df['Dx ID'][i]) == df_prev['id'][j]:
#            df['Dx Prevalence'][i] = df_prev['prev'][j]

df = pd.merge(df, df_prev[['Dx ID', 'Dx Prevalence']], on='Dx ID')
```

```python
df.head()

df.to_csv('mxdx.csv')
```

```python
# cast all numeric values to integers 

#for i in range(len(df)):  
#    df['Dx Prevalence'][i] = int(df['Dx Prevalence'][i])    
#    df['Mx PPV'][i] = int(df['Mx PPV'][i])    
#    df['Mx Sensitivity'][i] = int(df['Mx Sensitivity'][i])    
#    df['Mx Import'][i] = int(df['Mx Import'][i])    

# Grayson, here is a vectorized form that should be quicker
cols = ['Dx Prevalence', 'Mx PPV', 'Mx Sensitivity', 'Import Num']
df[cols] = df[cols].apply(pd.to_numeric)
df.head()
```

```python
#PREV = {1: 1/1000000,
#       2: 1/100000,
#       3: 1/1000,
#       4: 1/100,
#       5: 1/10}
PREV = {1: 1/1000000,
       2: 1/100000,
       3: 1/1000,
       4: 1/100,
       5: 1/15}

PPV = {0: np.NaN,
      1: 0.01,
      2: 0.1,
      3: 0.75,
      4: 0.9,
      5: 1}

SENS = {1: 0.01,
       2: 0.1,
       3: 0.5,
       4: 0.75,
       5: 1}

df['Dx Prevalence'] = df['Dx Prevalence'].map(PREV)
df['Mx PPV'] = df['Mx PPV'].map(PPV)
df['Mx Sensitivity'] = df['Mx Sensitivity'].map(SENS)
```

```python
df.head()
```

```python
# add specificity helper function

def calc_specificity(sens, ppv, prev):
    """
    Calculate specificity based on sensitivity, positive
    predictive value, & prevalence. 
    """
    spec = 1 - ((sens*prev/ppv - sens*prev)/(1-prev))
    return spec
```

```python
# calculate specificities

#for i in range(len(df)):
#    df['Specificity'][i] = calc_specificity(df['Mx Sensitivity'][i], df['Mx PPV'][i], df['Dx Prevalence'][i])

# Grayson, here is vectorized notation coming into play again
df['Specificity'] = calc_specificity(df['Mx Sensitivity'], df['Mx PPV'], df['Dx Prevalence'])
df.head()
```

```python
#for i in range(len(df)):
#    df['LR (+)'][i] = df['Mx Sensitivity'][i]/(1-df['Specificity'][i])
#    df['LR (-)'][i] = (1-df['Mx Sensitivity'][i])/df['Specificity'][i]

def lr_pos(sens, spec):
    return sens/(1-spec)

def lr_neg(sens, spec):
    return (1-sens)/spec

df['LR (+)'] = lr_pos(df['Mx Sensitivity'], df['Specificity'])
df['LR (-)'] = lr_neg(df['Mx Sensitivity'], df['Specificity'])
```
